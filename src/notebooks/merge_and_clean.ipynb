üìå Paso 1: Cargar los datos originales

```python
import pandas as pd

# Cargar los datasets
mat_path = '../data_sample/student-mat.csv'
por_path = '../data_sample/student-por.csv'

df_mat = pd.read_csv(mat_path, sep=';')
df_por = pd.read_csv(por_path, sep=';')

print("student-mat:", df_mat.shape)
print("student-por:", df_por.shape)
```

üîÅ Paso 2: Fusionar los datasets

```python
# Columnas clave para hacer merge
merge_cols = ["school", "sex", "age", "address", "famsize", "Pstatus",
              "Medu", "Fedu", "Mjob", "Fjob", "reason", "nursery", "internet"]

# Hacemos merge como en el script R
df_merged = pd.merge(df_mat, df_por, on=merge_cols, suffixes=('_mat', '_por'))

print("Dataset fusionado:", df_merged.shape)
df_merged.head()
```

üìä Paso 3: Inspecci√≥n b√°sica

```python
# Visualizamos las columnas y algunos datos
df_merged.columns.tolist()
df_merged.describe()
```

üíæ Paso 4: Guardar una muestra ligera para GitHub

```python
# Guardamos una muestra de 100 registros aleatorios
sample = df_merged.sample(100, random_state=42)

# Guardar como CSV
sample_path = '../data_sample/student-merged-sample.csv'
sample.to_csv(sample_path, index=False)

print(f"Muestra guardada en: {sample_path}")
```
